```python
j = 5
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

menor = float('inf')  # Initialize with infinity to ensure any first loss is smaller
lista = []
arquivo = np.load(f'teste{j}.npy')
x = arquivo[0]
y = np.ravel(arquivo[1])

# Initialize these arrays after loading x and y to get the correct shapes
xMin = np.empty_like(x)
yMin = np.empty_like(y)
y_estMin = np.empty_like(y)  # Assuming y_est will have the same shape as y
lossMin = None  # This will be updated later

for i in range(10):
    regr = MLPRegressor(hidden_layer_sizes=(8,4),
                        max_iter=10000,
                        activation='relu',
                        solver='adam',
                        learning_rate='adaptive',
                        n_iter_no_change=50)
    regr = regr.fit(x, y)
    y_est = regr.predict(x)
    loss = regr.best_loss_

    if loss < menor:
        menor = loss
        xMin = np.copy(x)
        yMin = np.copy(y)
        y_estMin = np.copy(y_est)
        lossMin = np.copy(regr.loss_curve_)

    lista.append(loss)

print(f"Arquivo teste {j}")
print(f"MÃ©dia do arquivo: {sum(lista) / len(lista)}")
print(f"Menor: {menor}")

plt.figure(figsize=[14,7])

# Plot original course
plt.subplot(1,3,1)
plt.plot(xMin, yMin)

# Plot learning curve
plt.subplot(1,3,2)
if lossMin is not None:
    plt.plot(lossMin)

# Plot regressor
plt.subplot(1,3,3)
plt.plot(xMin, yMin, linewidth=1, color='yellow')
plt.plot(xMin, y_estMin, linewidth=2)

plt.show()
```
